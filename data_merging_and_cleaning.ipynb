```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
```

College Data

```python
innovation = pd.read_csv('innovation_rates_by_college.csv')
innovation.rename(columns={
    'instnm':'college_name',
    'count': 'total_students',
    'count_pq1': 'lower_class',
    'count_pq2': 'lower_middle_class',
    'count_pq3': 'middle_class',
    'count_pq4': 'upper-middle_class',
    'count_pq5': 'upper_class',
    'inventor_pq1': 'lower_class_inventor',
    'inventor_pq2': 'lower-middle_class_inventor',
    'inventor_pq3': 'middle_class_inventor',
    'inventor_pq4': 'upper_middle_class_inventor',
    'inventor_pq5': 'upper_class_inventor',
    'top5cit': 'top5citedinventors'}, inplace=True)
innovation.head()
```

```python
baseline = pd.read_csv('baseline_cross-sec_estimates.csv')
baseline.rename(columns={
    'name':'college_name',
}, inplace=True)
baseline.head()
```

```python
merged_df = pd.merge(innovation, baseline, on='college_name', how='inner')
merged_df.info()

for column_name in merged_df.columns:
    print(column_name)


```

```python
drop = [17, 20, 23, 24, 25, 26, 27]
cleaned_df = merged_df.drop(merged_df.columns[drop], axis=1)
cleaned_df.info()
```

```python
cleaned_df.dropna(inplace = True)

cleaned_df['type'] = cleaned_df['type'].astype('category')
cleaned_df['tier'] = cleaned_df['tier'].astype('category')
cleaned_df['iclevel'] = cleaned_df['iclevel'].astype('category')
cleaned_df['region'] = cleaned_df['region'].astype('category')
cleaned_df['multi'] = cleaned_df['multi'].astype('bool')
cleaned_df['imputed'] = cleaned_df['imputed'].astype('bool')
cleaned_df.info()
```

```python
college = pd.read_csv('college_level_characteristics.csv')
college.head()
```

```python
college.info()
college.rename(columns={'name': 'college_name'}, inplace=True)

```

```python
merged_df2 = pd.merge(cleaned_df, college, on='college_name', how='inner')
merged_df.info()

for column_name in merged_df2.columns:
    print(column_name)
```

```python
drop = [ 94,95,96,97,98,99,100,101,102,103,104,105,106,111]
cleaned_df2 = merged_df2.drop(merged_df2.columns[drop], axis=1)
cleaned_df2.info()
```

```python
cleaned_df2.dropna(inplace = True)
cleaned_df2['barrons'] = cleaned_df2['barrons'].astype('category')
cleaned_df2['public'] = cleaned_df2['public'].astype('bool')
cleaned_df2['flagship'] = cleaned_df2['flagship'].astype('bool')
cleaned_df2['hbcu'] = cleaned_df2['hbcu'].astype('bool')
cleaned_df2.info()
```
